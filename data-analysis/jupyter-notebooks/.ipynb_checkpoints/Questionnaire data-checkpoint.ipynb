```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np
import matplotlib
import sys
```

## Load our data from file

```python
LocationAllData = "../data-cleanup/merged-cleaned-csv/driving_data_merged_1_to_20.csv"
df = pd.read_csv(LocationAllData)
```

## Create a new column with distance window
Each row will be assigned a "distance window". Each window will be averaged to resample our data

```python
# Decide which distance to average over
distanceWindowWidth = 100

def assignDistanceWindow(distance):
  # Get which distance pocket it would fit in.   
  distanceWindowIndex = round(distance/distanceWindowWidth)
  return distanceWindowWidth * distanceWindowIndex

# Add a new row which containts which distance group each row belongs to  
df["distanceWindow"] = df["distanceTraveled"].apply(assignDistanceWindow)
```

## Create a new dataframe with the data we need and group it appropriately
The grouping is based on how the average speed is supposed to be calculated. For each attempt basically.

```python
# Group the data based on attempt, which EVIS and userID
# Set as_index to false to allow access to these values later on for plotting
needed_data = df[["Attempt nr","evisID","userID", "distanceWindow","speed","yPosition"]].copy()

groups = needed_data.groupby(["Attempt nr","evisID","userID", "distanceWindow"], as_index=False)

# Compute mean values in each group and store the new dataframe
# Create a group object again from the dataframe to allow for easy itteration when plotting
average_speed_individual = groups.mean()
average_speed_individual_grouped = average_speed_individual.groupby(["Attempt nr","evisID","userID"])

# There will be a different number of samples averaged at longer distances, since everyone did not make it the enture 8k
# Calculate average speed for all users separated by attempt and EVSIS
# Finally re-group them based on how we want to plot them
average_speed_group_based_grouped = average_speed_individual.groupby(["Attempt nr","evisID","distanceWindow"], as_index=False)
average_speed_group_based = average_speed_group_based_grouped.mean()
average_speed_group_based_grouped = average_speed_group_based.groupby(["Attempt nr","evisID"])

# Calculate the average road height based on every attempt
road_height = needed_data[["distanceWindow","yPosition"]]
road_height = road_height.rename(columns={"yPosition":"roadHeight"})
average_road_height_grouped = road_height.groupby(["distanceWindow"], as_index=False)
average_road_height = average_road_height_grouped.mean()
```

## Customize our plot window

```python
plt.style.use("default")
params = {
  'font.size': 20, 
  'axes.labelsize': 18,
  'axes.titlesize':20, 
  'legend.fontsize': 18, 
  'xtick.labelsize': 18, 
  'ytick.labelsize': 18
 }
matplotlib.rcParams.update(params)
```

## Plot settings

```python
whichAttempt = 2
```

## Plot each attempt as transparent lines

```python
%%capture out
# Prevent displaying of plot in this cell

fig, ax =  plt.subplots(figsize=(15,10))
ax2 = ax.twinx()

ax.set_xlabel("Distance [m]")
ax.set_ylabel("Speed [km/h]")
ax.set_title("Speed over distance traveled during attempt #" + str(whichAttempt))



# Plot the data
for name,group in average_speed_individual_grouped:
    if name[0] == whichAttempt and name[1] == "GuessOMeter":
        ax.plot(group["distanceWindow"], group["speed"], color="C0", zorder=1, linewidth=2, alpha=0.4)
    if name[0] == whichAttempt and name[1] == "DiffAndCOPE1":
        ax.plot(group["distanceWindow"], group["speed"], color="C2", zorder=1, linestyle="-", linewidth=2, alpha=0.4)
```

## Mark the end of each line with a dot

```python
for name,group in average_speed_individual_grouped:
    if name[0] == whichAttempt and name[1] == "GuessOMeter":
        ax.plot(group["distanceWindow"].iloc[-1], group["speed"].iloc[-1], zorder=1, color="C0", marker="o", markersize=10, alpha=.5)
    if name[0] == whichAttempt and name[1] == "DiffAndCOPE1":
        ax.plot(group["distanceWindow"].iloc[-1], group["speed"].iloc[-1], zorder=1, color="C2", marker="o", markersize=10, alpha=.5)
```

## Plot clear average speed line for each EVIS group in attempt #2

```python
# Plot the data
for name,group in average_speed_group_based_grouped:
    if name[0] == whichAttempt and name[1] == "GuessOMeter":
        ax.plot(group["distanceWindow"], group["speed"], color="C0", zorder=3, linewidth=3, alpha=1)
    if name[0] == whichAttempt and name[1] == "DiffAndCOPE1":
        ax.plot(group["distanceWindow"], group["speed"], color="C2", zorder=3, linestyle="-", linewidth=3, alpha=1)
```

```python
color = 'k'
ax2.set_ylabel('Road height [m]', color=color)  # we already handled the x-label with ax1
ax2.plot(average_road_height["distanceWindow"], average_road_height["roadHeight"], color="k", 
          zorder=2, linewidth=2, linestyle="--", alpha=1)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
# Custom legend  
first_patch = patches.Patch(color="C0", label="guess-o-meter attempt #" + str(whichAttempt))
second_patch = patches.Patch(color="C2", label="diff+COPE1 attempt #" + str(whichAttempt))
third_patch = patches.Patch(color="k", label="Road height")

ax.legend(handles=[first_patch, second_patch, third_patch])

ax.set_ylim(0,120)
ax2.set_ylim(-10,30)

ax.set_yticks(np.arange(0,121,10)) 
#ax.set_yticks((0, 50, 100, 150)) 
#ax2.set_yticks(np.arange(-10,31,5)) 

ax.grid(True)
#ax2.grid(True)

ax.set_zorder(1)
ax.patch.set_visible(False)

fig.tight_layout()

fig
```
